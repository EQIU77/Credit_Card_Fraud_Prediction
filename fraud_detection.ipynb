```python
import numpy as np
fraud_ind = np.array(data[data.Class == 1].index)
normal_ind = data[data.Class == 0].index
n_fraud = len(data[data.Class == 1])
# random selection from normal class
random_normal_ind = np.random.choice(normal_ind, n_fraud, replace = False)
random_normal_ind = np.array(random_normal_ind)
# merge two class indices: random genuine + original fraud
under_sample_ind = np.concatenate([fraud_ind,random_normal_ind])
# Under sample dataset
undersample_data = data.iloc[under_sample_ind,:]
y_undersample  = undersample_data['Class'].values #target
X_undersample = undersample_data.drop(['Class'],axis=1).values #features

print("# transactions in undersampled data: ", len(undersample_data))
print("% normal transactions: ",len(undersample_data[undersample_data.Class == 0])/len(undersample_data))
print("% fraud transactions: ", sum(y_undersample)/len(undersample_data))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_undersample, y_undersample)

X_train_new = scalar.fit_transform(X_train_new)
X_test_new = scalar.transform(X_test_new)
```

```python
from sklearn import linear_model
model_lr = linear_model.LogisticRegression(random_state=1)
```

```python
model_lr.fit(X_train_new, y_train_new)
pred = model_lr.predict(X_test_new)
pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_new, pred)
print(cm)
```

```python
print("Logistic regression's accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
print("Logistic regression's recall", (cm[1,1])/(cm[1,1]+cm[1,0]))
print("Logistic regression's precision", (cm[1,1])/(cm[1,1]+cm[0,1]))
```

```python
from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import plot_roc_curve
plot_precision_recall_curve(model_lr,X_test_new,y_test_new)
plot_roc_curve(model_lr,X_test_new,y_test_new)
```

```python
from sklearn import ensemble
model_rf = ensemble.RandomForestClassifier(random_state=1)
model_rf.fit(X_train_new, y_train_new)
pred= model_rf.predict(X_test_new)
cm = confusion_matrix(y_test_new, pred)
print(cm)
print("Random forest's accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
print("Random forest's recall", (cm[1,1])/(cm[1,1]+cm[1,0]))
print("Random forest's precision", (cm[1,1])/(cm[1,1]+cm[0,1]))
plot_precision_recall_curve(model_rf,X_test_new,y_test_new)
plot_roc_curve(model_rf,X_test_new,y_test_new)
```

```python
from xgboost import XGBClassifier

model_xgb = XGBClassifier(random_state = 1)
model_xgb.fit(X_train_new, y_train_new)
pred= model_xgb.predict(X_test_new)
cm = confusion_matrix(y_test_new, pred)
print(cm)
print("XGBoost's accuracy is ", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
print("XGBoost's recall is", (cm[1,1])/(cm[1,1]+cm[1,0]))
print("XGBoost's precision is", (cm[1,1])/(cm[1,1]+cm[0,1]))
plot_precision_recall_curve(model_xgb,X_test_new,y_test_new)
plot_roc_curve(model_xgb,X_test_new,y_test_new)
```
